```python
# Define your figure and subplots
fig = plt.figure(constrained_layout=True, figsize=(12, 4))
gs = GridSpec(1, 3, figure=fig)
ax = fig.add_subplot(gs[0, 0])
ax_pie = fig.add_subplot(gs[0, 1])

# Donut Plot - Customer Type by Segment
colors = ['skyblue', 'cornflowerblue', 'silver']
label_names = df['Segment'].value_counts().index

# Assuming you have the count of each segment in the DataFrame, segment_counts
segment_counts = df['Segment'].value_counts()

# Plotting bar chart
ax.bar(label_names, segment_counts, color=colors)
ax.set_title('Customer Type by Segment', color='purple', size=14)
ax.set_xlabel('Segment')
ax.set_ylabel('Count')


# Plotting pie chart
ax_pie.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
ax_pie.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax_pie.set_title('Customer Type Distribution', size=14, color='purple')

plt.show()
```

```python
# Building a figure
fig = plt.figure(constrained_layout=True, figsize=(17, 5))

# Axis definition with GridSpec
gs = GridSpec(1, 3, figure=fig)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1:2])

# Grouping data by Segment with Sales
df_segment_sales = df.groupby(by='Segment', as_index=False)['Sales'].sum()  # Summing sales per segment
df_segment_sales = df_segment_sales.sort_values(by='Sales', ascending=False).reset_index(drop=True)

# Most Sales by Segment
sns.barplot(y='Segment', x='Sales', data=df_segment_sales, ax=ax1, hue='Segment', palette='magma', legend=False)
AnnotateBars(n_dec=0, font_size=10, color='black').horizontal(ax1)
format_spines(ax1, right_border=False)
ax1.set_title('Most Sales by Segment', size=18, color='dimgrey')
ax1.set_ylabel('Segment Type')

# Grouping data by Segment with Quantity
df_segment_quantity = df.groupby(by='Segment', as_index=False)['Quantity'].sum()  # Summing quantity per segment

# Most Quantity of Item Ordered by Segment
sns.barplot(y='Quantity', x='Segment', data=df_segment_quantity, ax=ax2, hue='Segment', palette='magma', legend=False)
format_spines(ax2, right_border=False)
ax2.set_title('Most Quantity of Item Ordered by Segment', size=18, color='dimgrey')
ax2.set_ylabel('Quantity')

plt.show()
```
